```python
import time
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

sentences = ["frase exemplo"] * 1000

for batch_size in [1, 8, 16, 32, 64, 128]:
    start = time.time()
    model.encode(sentences, batch_size=batch_size)
    end = time.time()

    total_time = end - start
    throughput = len(sentences) / total_time

    print(f"Batch: {batch_size}")
    print(f"Tempo: {total_time:.2f}s")
    print(f"Throughput: {throughput:.2f} frases/s")
    print("-----")
```
